Word Machine
A word machine is a system that performs a sequence of simple operations on a stack of integers.
Initially, the stack is empty. The sequence of operations is given as a string. Operations are separated by single spaces. The following operations may be specified:

• An integer X (from 0 to 220 - 1): the machine pushes X onto the stack;
• "POP": the machine removes the topmost number from the stack;
• "DUP": the machine pushes a duplicate of the topmost number onto the stack;
• "+": the machine pops the two topmost elements from the stack, adds them together and pushes the sum onto the stack;
• "-": the machine pops the two topmost elements from the stack, subtracts the second one from the first (topmost) one, and pushes the difference onto the stack.

After processing all the operations, the machine returns the topmost value from the stack.
The machine processes 20-bit unsigned integers (numbers from 0 to 220 - 1). An overflow in addition or underflow in subtraction causes an error. The machine also reports an error when it tries to perform an operation that expects more numbers on the stack than the stack actually contains.

Also, if after performing all the operations, the stack is empty, the machine reports an error.
For example, given a string "4 5 6 - 7 +", the machine performs the following operations:

Operation  |       Comment         |   stack
           |                       |    [empty]
"4"        |    push 4             |               
           |                       |      4
"5"        |    push 5             |            
           |                       |      4, 5
"6"        |    push 6             | 
           |                       |      4, 5, 6
"-"        |   subtract 5 from 6   |        
           |                       |      4,1
"7"        |    push 7             | 
           |                       |      4, 1, 7
"+"        |    add 1 and 7        |        
           |                       |      4, 8

Finally, the machine will return 8.

Given a string "13 DUP 4 POP 5 DUP + DUP + -", the machine performs the following operations:
Operation  |       Comment         |   stack
           |                       |    [empty]
"13"       |    push 13            |    
           |                       |    13
"DUP"      |    duplicate 13       |    
           |                       |    13, 13
"4"        |    push 4             |    
           |                       |    13, 13, 4
"POP"      |    pop 4              |    
           |                       |    13, 13
"5"        |    push 5             |    
           |                       |    13, 13, 5
"DUP"      |    duplicate 5        |    
           |                       |    13, 13, 5, 5
"+"        |    add 5 and 5        |    
           |                       |    13, 13, 10
"DUP"      |    duplicate 10       |    
           |                       |    13, 13, 10, 10
"+"        |    add 10 and 10      |    
           |                       |    13, 13, 20
"-"        |   subtract 13 from 20 |    
           |                       |    13, 7
Finally, the machine will return 7.

```python
def word_machine(operaciones):
    # Crear el stack vacío
    stack = []
    
    # Dividir las operaciones en una lista de caracteres
    caracteres = operaciones.split()
    
    # Procesar cada carácter
    for caracter in caracteres:
        if caracter.isdigit():  # Si es un número
            numero = int(caracter)
            # **Validación del rango permitido para los números**
            if 0 <= numero <= (2**20 - 1):
                stack.append(numero)  # Agregar el número al stack
                print(f"Número {numero} añadido al stack: {stack}")
            else:
                print(f"Error: Número {numero} fuera del rango permitido.")
                return -1
        
        elif caracter == "POP":  # Operación POP
            if stack:  # Validar si la pila no está vacía
                eliminado = stack.pop()  # Eliminamos el último número
                print(f"POP ejecutado, se eliminó {eliminado}. Stack actual: {stack}")
            else:
                print("Error: No se puede hacer POP en un stack vacío.")
                return -1
        
        elif caracter == "DUP":  # Operación DUP
            if stack:  # Validar si la pila no está vacía
                duplicado = stack[-1]
                stack.append(duplicado)
                print(f"DUP ejecutado, se duplicó {duplicado}. Stack actual: {stack}")
            else:
                print("Error: No se puede hacer DUP en un stack vacío.")
                return -1
        
        elif caracter == "+":  # Operación +
            if len(stack) >= 2:  # Validar que haya al menos dos elementos
                a = stack.pop()
                b = stack.pop()
                resultado = a + b
                # **Validación de Overflow**
                if resultado <= (2**20 - 1):  # Si está dentro del rango permitido
                    stack.append(resultado)
                    print(f"Suma ejecutada: {b} + {a} = {resultado}. Stack actual: {stack}")
                else:
                    print(f"Error: Overflow en la suma {b} + {a} = {resultado}.")
                    return -1
            else:
                print("Error: No hay suficientes números en el stack para sumar.")
                return -1
        
        elif caracter == "-":  # Operación -
            if len(stack) >= 2:  # Validar que haya al menos dos elementos
                a = stack.pop()
                b = stack.pop()
                resultado = b - a
                # **Validación de Underflow**
                if resultado >= 0:  # Si no es negativo
                    stack.append(resultado)
                    print(f"Resta ejecutada: {b} - {a} = {resultado}. Stack actual: {stack}")
                else:
                    print(f"Error: Underflow en la resta {b} - {a} = {resultado}.")
                    return -1
            else:
                print("Error: No hay suficientes números en el stack para restar.")
                return -1

    # **Modificación: Presentar solo el último elemento si el stack tiene más de uno**

    print(f"Stack antes de aplicar la condición final: {stack}")
    if len(stack) > 1:  # Si hay más de un número en el stack
        resultado = stack[-1]  # Tomamos el último elemento
        print(f"Condición aplicada: Stack tiene más de un número, resultado final: {resultado}")
        return resultado
    elif len(stack) == 1:  # Si solo hay un número, retornamos ese número
        print(f"Stack tiene exactamente un número, resultado final: {stack[0]}")
        return stack[0]
    else:  # Si el stack está vacío
        print("Error: Stack vacío al finalizar las operaciones.")
        return -1

# Casos de prueba
S1 = "1048575 1048575 +"  # Caso de Overflow
S2 = "5 10 -"            # Caso de Underflow
S3 = "3 3 +"             # Caso válido
S4 = "1048575 1048578 -"       # Caso válido, resultado 0

print(f"Resultado para S1: {word_machine(S1)}")  # Esperado: -1 (Overflow)
print(f"Resultado para S2: {word_machine(S2)}")  # Esperado: -1 (Underflow)
print(f"Resultado para S3: {word_machine(S3)}")  # Esperado: 6
print(f"Resultado para S4: {word_machine(S4)}")  # Esperado: 0
```
