Word Machine
A word machine is a system that performs a sequence of simple operations on a stack of integers.
Initially, the stack is empty. The sequence of operations is given as a string. Operations are separated by single spaces. The following operations may be specified:

‚Ä¢ An integer X (from 0 to 220 - 1): the machine pushes X onto the stack;
‚Ä¢ "POP": the machine removes the topmost number from the stack;
‚Ä¢ "DUP": the machine pushes a duplicate of the topmost number onto the stack;
‚Ä¢ "+": the machine pops the two topmost elements from the stack, adds them together and pushes the sum onto the stack;
‚Ä¢ "-": the machine pops the two topmost elements from the stack, subtracts the second one from the first (topmost) one, and pushes the difference onto the stack.

After processing all the operations, the machine returns the topmost value from the stack.
The machine processes 20-bit unsigned integers (numbers from 0 to 220 - 1). An overflow in addition or underflow in subtraction causes an error. The machine also reports an error when it tries to perform an operation that expects more numbers on the stack than the stack actually contains.

Also, if after performing all the operations, the stack is empty, the machine reports an error.
For example, given a string "4 5 6 - 7 +", the machine performs the following operations:

```python
def word_machine(operaciones):
    # Crear el stack vac√≠o
    stack = []
    
    # Dividir las operaciones en una lista de caracteres
    caracteres = operaciones.split()
    
    # Procesar cada car√°cter
    for caracter in caracteres:
        if caracter.isdigit():  # Si es un n√∫mero
            numero = int(caracter)
            # **Validaci√≥n del rango permitido para los n√∫meros**
            if 0 <= numero <= (2**20 - 1):
                stack.append(numero)  # Agregar el n√∫mero al stack
                print(f"N√∫mero {numero} a√±adido al stack: {stack}")
        elif caracter == "POP":  # Operaci√≥n POP
            eliminado = stack.pop()  # Eliminamos el √∫ltimo n√∫mero
            print(f"POP ejecutado, se elimin√≥ {eliminado}. Stack actual: {stack}")
        elif caracter == "DUP":  # Operaci√≥n DUP
            duplicado = stack[-1]  # Duplicamos el √∫ltimo n√∫mero
            stack.append(duplicado)
            print(f"DUP ejecutado, se duplic√≥ {duplicado}. Stack actual: {stack}")
        elif caracter == "+":  # Operaci√≥n +
            a = stack.pop()
            b = stack.pop()
            resultado = a + b
            stack.append(resultado)  # Sumar los dos √∫ltimos n√∫meros
            print(f"Suma ejecutada: {b} + {a} = {resultado}. Stack actual: {stack}")
        elif caracter == "-":  # Operaci√≥n -
            a = stack.pop()
            b = stack.pop()
            resultado = b - a
            stack.append(resultado)  # Restar los dos √∫ltimos n√∫meros
            print(f"Resta ejecutada: {b} - {a} = {resultado}. Stack actual: {stack}")
        else:
            print(f"Operaci√≥n '{caracter}' no implementada todav√≠a. Stack actual: {stack}")
    
    # Retornar el stack final
    print(f"Stack final: {stack}")
    return stack

# Prueba con las operaciones completas
S = "1 2 POP 3 4 DUP - 5 +"
stack_final = word_machine(S)
print(f"Resultado para S = '{S}': {stack_final}")
```

##¬†Paso 3 Agregar los errores de overflow + underflow

```python
def word_machine(operaciones):
    # Crear el stack vac√≠o
    stack = []
    
    # Dividir las operaciones en una lista de caracteres
    caracteres = operaciones.split()
    
    # Procesar cada car√°cter
    for caracter in caracteres:
        if caracter.isdigit():  # Si es un n√∫mero
            numero = int(caracter)
            # **Validaci√≥n del rango permitido para los n√∫meros**
            if 0 <= numero <= (2**20 - 1):
                stack.append(numero)  # Agregar el n√∫mero al stack
                print(f"N√∫mero {numero} a√±adido al stack: {stack}")
            else:
                print(f"Error: N√∫mero {numero} fuera del rango permitido.")
                return -1
        
        elif caracter == "POP":  # Operaci√≥n POP
            if stack:  # Validar si la pila no est√° vac√≠a
                eliminado = stack.pop()  # Eliminamos el √∫ltimo n√∫mero
                print(f"POP ejecutado, se elimin√≥ {eliminado}. Stack actual: {stack}")
            else:
                print("Error: No se puede hacer POP en un stack vac√≠o.")
                return -1
        
        elif caracter == "DUP":  # Operaci√≥n DUP
            if stack:  # Validar si la pila no est√° vac√≠a
                duplicado = stack[-1]
                stack.append(duplicado)
                print(f"DUP ejecutado, se duplic√≥ {duplicado}. Stack actual: {stack}")
            else:
                print("Error: No se puede hacer DUP en un stack vac√≠o.")
                return -1
        
        elif caracter == "+":  # Operaci√≥n +
            if len(stack) >= 2:  # Validar que haya al menos dos elementos
                a = stack.pop()
                b = stack.pop()
                resultado = a + b
           ################# **Validaci√≥n de Overflow**   ##################
                if resultado <= (2**20 - 1):  # Si est√° dentro del rango permitido
                    stack.append(resultado)
                    print(f"Suma ejecutada: {b} + {a} = {resultado}. Stack actual: {stack}")
                else:
                    print(f"Error: Overflow en la suma {b} + {a} = {resultado}.")
                    return -1
            else:
                print("Error: No hay suficientes n√∫meros en el stack para sumar.")
                return -1
        
        elif caracter == "-":  # Operaci√≥n -
            if len(stack) >= 2:  # Validar que haya al menos dos elementos
                a = stack.pop()
                b = stack.pop()
                resultado = b - a
                ################# **Validaci√≥n de Underflow**  #################
                if resultado >= 0:  # Si no es negativo
                    stack.append(resultado)
                    print(f"Resta ejecutada: {b} - {a} = {resultado}. Stack actual: {stack}")
                else:
                    print(f"Error: Underflow en la resta {b} - {a} = {resultado}.")
                    return -1
            else:
                print("Error: No hay suficientes n√∫meros en el stack para restar.")
                return -1

    # **Modificaci√≥n: Presentar solo el √∫ltimo elemento si el stack tiene m√°s de uno**

    print(f"Stack antes de aplicar la condici√≥n final: {stack}")
    if len(stack) > 1:  # Si hay m√°s de un n√∫mero en el stack
        resultado = stack[-1]  # Tomamos el √∫ltimo elemento
        print(f"Condici√≥n aplicada: Stack tiene m√°s de un n√∫mero, resultado final: {resultado}")
        return resultado
    elif len(stack) == 1:  # Si solo hay un n√∫mero, retornamos ese n√∫mero
        print(f"Stack tiene exactamente un n√∫mero, resultado final: {stack[0]}")
        return stack[0]
    else:  # Si el stack est√° vac√≠o
        print("Error: Stack vac√≠o al finalizar las operaciones.")
        return -1

# Casos de prueba
S1 = "1048575 1048575 +"  # Caso de Overflow
S2 = "5 10 -"            # Caso de Underflow
S3 = "3 3 +"             # Caso v√°lido
S4 = "1048575 1048578 -"       # Caso v√°lido, resultado 0

print(f"Resultado para S1: {word_machine(S1)}")  # Esperado: -1 (Overflow)
print(f"Resultado para S2: {word_machine(S2)}")  # Esperado: -1 (Underflow)
print(f"Resultado para S3: {word_machine(S3)}")  # Esperado: 6
print(f"Resultado para S4: {word_machine(S4)}")  # Esperado: 0
```

## Paso 4 Poner las √∫ltimos posibles errores

- pila vacia, no queda numeros al finalizar la pila

```python
def word_machine(operaciones):
    # Crear el stack vac√≠o
    stack = []
    
    # Dividir las operaciones en una lista de caracteres
    caracteres = operaciones.split()
    
    # Procesar cada car√°cter
    for caracter in caracteres:
        if caracter.isdigit():  # Si es un n√∫mero
            numero = int(caracter)
            # Validaci√≥n del rango permitido para los n√∫meros
            if 0 <= numero <= (2**20 - 1):
                stack.append(numero)  # Agregar el n√∫mero al stack
                print(f"N√∫mero {numero} a√±adido al stack: {stack}")
            else:
                print(f"Error: N√∫mero {numero} fuera del rango permitido.")
                return -1
        
        elif caracter == "POP":  # Operaci√≥n POP
            # **Validaci√≥n: N√∫meros insuficientes para POP**
            if stack:
                eliminado = stack.pop()
                print(f"POP ejecutado, se elimin√≥ {eliminado}. Stack actual: {stack}")
            else:
                print("Error: No se puede hacer POP en un stack vac√≠o.")
                return -1
        
        elif caracter == "DUP":  # Operaci√≥n DUP
            # **Validaci√≥n: N√∫meros insuficientes para DUP**
            if stack:
                duplicado = stack[-1]
                stack.append(duplicado)
                print(f"DUP ejecutado, se duplic√≥ {duplicado}. Stack actual: {stack}")
            else:
                print("Error: No se puede hacer DUP en un stack vac√≠o.")
                return -1
        
        elif caracter == "+":  # Operaci√≥n +
            # **Validaci√≥n: N√∫meros insuficientes para suma**
            if len(stack) >= 2:
                a = stack.pop()
                b = stack.pop()
                resultado = a + b
                if resultado <= (2**20 - 1):  # Validaci√≥n de overflow
                    stack.append(resultado)
                    print(f"Suma ejecutada: {b} + {a} = {resultado}. Stack actual: {stack}")
                else:
                    print(f"Error: Overflow en la suma {b} + {a} = {resultado}.")
                    return -1
            else:
                print("Error: No hay suficientes n√∫meros en el stack para sumar.")
                return -1
        
        elif caracter == "-":  # Operaci√≥n -
            # **Validaci√≥n: N√∫meros insuficientes para resta**
            if len(stack) >= 2:
                a = stack.pop()
                b = stack.pop()
                resultado = b - a
                if resultado >= 0:  # Validaci√≥n de underflow
                    stack.append(resultado)
                    print(f"Resta ejecutada: {b} - {a} = {resultado}. Stack actual: {stack}")
                else:
                    print(f"Error: Underflow en la resta {b} - {a} = {resultado}.")
                    return -1
            else:
                print("Error: No hay suficientes n√∫meros en el stack para restar.")
                return -1

    # **Validaci√≥n: Pila vac√≠a al finalizar**
    if not stack:
        print("Error: Stack vac√≠o al finalizar las operaciones.")
        return -1

    # Si hay m√°s de un n√∫mero, devolver el √∫ltimo n√∫mero sin error
    print(f"Stack antes de finalizar: {stack}")
    resultado = stack[-1]  # Devolver el √∫ltimo n√∫mero de la pila
    print(f"Resultado final: {resultado}")
    return resultado

# Casos de prueba
S1 = "POP"                 # Error: POP en un stack vac√≠o
S2 = "3 +"                 # Error: N√∫meros insuficientes para suma
S3 = "1 2 +"               # Caso v√°lido
S4 = "3 DUP -"             # Error: Underflow en la resta
S5 = ""                    # Error: Pila vac√≠a al finalizar

print(f"Resultado para S1: {word_machine(S1)}")  # Esperado: -1
print(f"Resultado para S2: {word_machine(S2)}")  # Esperado: -1
print(f"Resultado para S3: {word_machine(S3)}")  # Esperado: 3
print(f"Resultado para S4: {word_machine(S4)}")  # Esperado: -1
print(f"Resultado para S5: {word_machine(S5)}")  # Esperado: -1
```

```python
def solution(S):
    # **Restricci√≥n 1: Validar el tama√±o de la cadena**
    if not (1 <= len(S) <= 100000):
        print("Error: La longitud de la cadena S est√° fuera del rango permitido.")
        return -1  # Indicamos un error con un valor especial
    
    # **Restricci√≥n 2: Validar que la cadena solo contenga d√≠gitos**
    if not S.isdigit():
        print("Error: La cadena S contiene caracteres no v√°lidos.")
        return -1  # Indicamos un error con un valor especial
    
    # Si las validaciones pasan, imprimimos la cadena recibida
    print(f"Cadena v√°lida recibida: {S}")
```
